```python
import numpy
```

```python
x = numpy.arange(10,21,2) # Array de enteros con paso 2
print(x)

y = numpy.sin(x)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.plot(x)
plt.ylabel('Valores de X')
plt.show()
```

```python
plt.figure(2)
plt.plot(y)
plt.ylabel('Valores de Y')
```

```python
plt.figure(3)
plt.plot(x,y)
plt.xlabel('Seno(x)')
plt.ylabel('X')
```

```python
# Ahora las dos a la vez:

z = numpy.cos(x)

plt.figure(4)
plt.plot(x,y,'r')
plt.plot(x,z,'g')

plt.legend(['Seno(x)', 'Coseno(x)'])
plt.xlabel('X')
plt.ylabel('Valores')
```
