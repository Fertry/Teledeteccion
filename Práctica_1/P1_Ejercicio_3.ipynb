```python
# Librerías de tratamiento de imágenes:
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Leer imágenes:
img = cv2.imread('img.png')
print(img)
```

```python
'''
Estructura de datos de la variable previa.

La estructura que emplea OpenCV para almacenar imágenes es un
array de matrices. Mediante la clase 'Mat' de la librería se
puede ver como se guarda una cabecera y un puntero; dónde la
cabecera contiene la información relativa al tamaño, método y
dirección de la matriz; y el puntero guarda los valores de 
cada pixel de la imagen.
'''
```

```python
plt.imshow(img)
#plt.show()
```

```python
# Editar una imagen:
print("Imagen de tamaño: ", img.shape)
print("Pixel concreto (100,100): ", img[100,100])
print("Fila concreta (100): ", img[100, :])
print("Subimagen de la imagen original: ", img[100:200, 100:200])
```

```python
# Modificaciones a la imagen:
b,g,r = cv2.split(img) # Extraer los colores

img_rgb = cv2.merge((r,g,b))
plt.imshow(img_rgb)
```

```python
# Extraer trozos concretos de la imagen:
trozo = img[0:50, 50:100]
lenx, leny, ch = trozo.shape

img[10:10+lenx, 15:15+leny] = trozo
img[20:20+lenx, 25:25+leny] = trozo
img[30:30+lenx, 40:40+leny] = trozo

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Guardar 
cv2.imwrite("Resultado.png", img)
```
