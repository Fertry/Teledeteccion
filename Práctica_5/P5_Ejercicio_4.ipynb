```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Imagen propia asignada; en mi caso la 10
imagen = cv.imread('img/10_NorthCarolina.jpg',cv.IMREAD_GRAYSCALE)
```

```python
# Función densi
def densi(imagen):
    # Resultado original: matriz de ceros del tamaño de entrada:
    resultado = np.zeros((imagen.shape[0], imagen.shape[1], 3), np.uint8)
    
    # Recorrer la imagen; en rangos aleatorios asignar colores random:
    for i in range(resultado.shape[0]):
        for j in range(resultado.shape[1]):
            
            # Creo 4 rangos aleatorios: 0-64 64-128 128-192 192-255
            if 0 < imagen[i,j] < 64:
                # Asigno valores aleatorios:
                resultado[i,j] = [0,64,64]
            elif 64 < imagen[i,j] < 128:
                # Asigno valores aleatorios:
                resultado[i,j] = [128,0,128]
            elif 128 < imagen[i,j] < 192:
                # Asigno valores aleatorios:
                resultado[i,j] = [192,192,0]
            else:
                # Asigno valores aleatorios:
                resultado[i,j] = [255,255,255]
            
    return resultado
```

```python
# Prueba con valores aleatorios (esto no se pide)
def densiPrueba(imagen):
    # Resultado original: matriz de ceros del tamaño de entrada:
    resultado = np.zeros((imagen.shape[0], imagen.shape[1], 3), np.uint8)
    
    # Recorrer la imagen; en rangos aleatorios asignar colores random:
    for i in range(resultado.shape[0]):
        for j in range(resultado.shape[1]):
            
            # Creo 4 rangos aleatorios: 0-64 64-128 128-192 192-255
            if 0 < imagen[i,j] < 64:
                # Creo números aleatorios para asignarlos:
                G = random.randint(1,255)
                B = random.randint(1,255)
                resultado[i,j] = [0,G,B]
            elif 64 < imagen[i,j] < 128:
                # Creo números aleatorios para asignarlos:
                R = random.randint(1,255)
                B = random.randint(1,255)
                resultado[i,j] = [R,0,B]
            elif 128 < imagen[i,j] < 192:
                # Creo números aleatorios para asignarlos:
                R = random.randint(1,255)
                G = random.randint(1,255)
                resultado[i,j] = [R,G,0]
            else:
                # Creo números aleatorios para asignarlos:
                R = random.randint(1,255)
                G = random.randint(1,255)
                B = random.randint(1,255)
                resultado[i,j] = [R,G,B]
            
    return resultado
```

```python
# Imagen original:
plt.imshow(imagen)
```

```python
# Density slicing aplicado: OJO esto tarda un rato!
resultado = densi(imagen)
plt.imshow(resultado)
cv.imwrite("salida/4_DensitySlicing.jpg", resultado)
```

```python
# Esto es una prueba con valores totalmente aleatorios! Esto tambien tarda lo suyo....
resultadoPrueba = densiPrueba(imagen)
plt.imshow(resultadoPrueba)
```
