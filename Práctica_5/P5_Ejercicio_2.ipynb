```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Imagen propia asignada; en mi caso la 10
imagen = cv.imread('img/10_NorthCarolina.jpg',cv.IMREAD_GRAYSCALE)
```

```python
# Función seudo (cambio min/max por un valor entero)
def seudo(imagen, corte):
    # Copiar la imagen no funciona si luego se setea un valor como [r,g,b] 
    # así que uso una matriz de ceros como el ejercicio anterior:
    resultado = np.zeros((imagen.shape[0], imagen.shape[1], 3), np.uint8)
    for i in range(imagen.shape[0]):
        for j in range(imagen.shape[1]):
            if imagen[i,j] < corte:
                # Destacamos el azul
                resultado[i,j] = [0,0,255]
            elif imagen[i,j] >= corte:
                # Destacamos el rojo
                resultado[i,j] = [255,0,0]

    return resultado
```

```python
# Imagen original:
plt.imshow(imagen)
```

```python
# Aplicar la función con corte de 128 (tarda un poco)
resultado = seudo(imagen, 128)
plt.imshow(resultado)
cv.imwrite("salida/2_CortadoI.jpg", resultado)
```

```python
# En el laboratorio se vio otra imagen más apropiada:
prueba = cv.imread('img/satelite_nocturno.jpg',cv.IMREAD_GRAYSCALE)
resultado_prueba = seudo(prueba, 128)
plt.imshow(resultado_prueba)
cv.imwrite("salida/2_CortadoII.jpg", resultado_prueba)
```
