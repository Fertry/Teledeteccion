```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv.imread('img/10_NorthCarolina.jpg',cv.IMREAD_GRAYSCALE)
```

```python
# Matriz de ejemplo del enunciado:
matriz_ejemplo = np.array([
    [0,1,3,2,4,3],
    [2,4,2,1,2,2],
    [7,8,5,6,4,3],
    [4,9,8,5,5,5],
    [4,8,7,5,4,5],
    [4,6,7,5,4,6]
])
```

```python
# Función diezmado: imagen + factor de entrada.
def diezmado(img, R):
     
    valores_digitales = []
    
    # Leer cada pixel pero nos quedamos solo con ciertos
    # valores digitales en base a (1/R)
    i = 0
    while i < img.shape[0]:
        j = 0
        while j < img.shape[1]:
            valores_digitales.append(img[i,j])
            j += int(1/R)
        i += int(1/R)
    
    # 'reshape' transforma esa lista en imagen:
    return np.reshape(valores_digitales,
                        [int(np.ceil(img.shape[0] / (int(1/R)))),
                        int(np.ceil(img.shape[1] / (int(1/R))))])
```

```python
print("Matriz de ejemplo: \n" + str(matriz_ejemplo))
print("Matriz de ejemplo diezmada: \n" + str(diezmado(matriz_ejemplo, 0.5)))
```

```python
# Aplicado a la imagen: se puede ver la diferencia en la escala (los nºs cambian!)
print("Original")
plt.imshow(imagen,cmap='gray')
```

```python
print("Diezmado al 0.5")
imagen_05 = diezmado(imagen, 0.5)
plt.imshow(imagen_05,cmap='gray')
cv.imwrite("salida/Diezmado_50.png", imagen_05)
```

```python
print("Diezmado al 0.33")
imagen_033 =diezmado(imagen, 0.33)
plt.imshow(imagen_033,cmap='gray')
cv.imwrite("salida/Diezmado_33.png", imagen_033)
```

```python
print("Diezmado al 0.25")
imagen_025 =diezmado(imagen, 0.25)
plt.imshow(imagen_025,cmap='gray')
cv.imwrite("salida/Diezmado_25.png", imagen_025)
```
