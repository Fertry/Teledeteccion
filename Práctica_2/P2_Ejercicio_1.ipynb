```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer la imagen (monobanda):
img = cv2.imread('10_NorthCarolina.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Función histograma del ejercicio mediante OpenCV:
def histograma(img):
    var = 0   
    v = [0]*256
    print("La dimension de la imagen es: " + str(img.shape))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # Toma el valor de ese pixel y actualiza el array:
            pixel = img[i][j]
            v[pixel] += 1
            var += 1
    # Genera el bar con una lista de 256 valores y sus frecuencias:
    plt.bar(np.arange(0, 256, 1), v)
    plt.show()
```

```python
'''
Otra forma de hacer esto mucho más rápido es con OpenCV
'''
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
# Llamo a la función: cuidado que tarda en cargar unos segundos!
histograma(img)
```
